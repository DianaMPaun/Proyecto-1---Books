```python
import matplotlib.pyplot as plt
import pandas as pd
```

1. How many book read in 2022?

```python
#Books read in 2022

books_read = {"Books read" : ["more than 20 books", "11-20 books", "6-10 books", "1-5 books", "No books"],
              "Percentage" :  ["27%", "19%", "18%", "32%", "4%"],}
books_read_stat = pd.DataFrame(books_read)
books_read_stat.index = books_read_stat.index + 1
books_read_stat.head()
```

```python

books_read_stat['Percentage'] = books_read_stat['Percentage'].str.rstrip("%").astype(float)

# Create a bar plot
plt.figure(figsize=(8, 5))
plt.bar(books_read_stat['Books read'], books_read_stat['Percentage'], color='skyblue')

# Add title and labels
plt.title('Percentage of People Based on Number of Books Read')
plt.xlabel('Books read')
plt.ylabel('Percentage')

# Rotate x-axis labels for better readability
plt.xticks(rotation=0, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

The survey says: According to the latest book reading statistics, 27% said they have read more than 20 books in 2023,
while 32% said they have read one to five books. Only 18% said they have read six to 10 books, and 19% said 11 to 20 books in 2023.
